```
from scipy.stats import chi2_contingency
import numpy as np
import pandas as pd


def cramers_V(var1,var2) :
    """
    calculate Cramers V statistic for categorial-categorial association for 2 variables
    """
    crosstab =np.array(pd.crosstab(var1,var2, rownames=None, colnames=None)) # Cross table building
    stat = chi2_contingency(crosstab)[0] # Keeping of the test statistic of the Chi2 test
    obs = np.sum(crosstab) # Number of observations
    mini = min(crosstab.shape)-1 # Take the minimum value between the columns and the rows of the cross table
    return (stat/(obs*mini))


def CramersCorrected(x,y):

    """ calculate Cramers V statistic for categorial-categorial association.
        uses correction from Bergsma and Wicher, 
        Journal of the Korean Statistical Society 42 (2013): 323-328
    """
    
    result=-1
    # check if there is a constant value
    if len(x.value_counts())==1 :
        print("First variable is constant")
    elif len(y.value_counts())==1:
        print("Second variable is constant")
    # calcul du tableau crois
    else:   
        conf_matrix=pd.crosstab(x, y)

        if conf_matrix.shape[0]==2:
            correct=False
        else:
            correct=True
            
        # Compute chi2
        chi2 = chi2_contingency(conf_matrix, correction=correct)[0]
        
        # Compute Cramer's V
        n = sum(conf_matrix.sum())
        phi2 = chi2/n
        r,k = conf_matrix.shape
        phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))    
        rcorr = r - ((r-1)**2)/(n-1)
        kcorr = k - ((k-1)**2)/(n-1)
        try:
            result=np.sqrt(phi2corr / min( (kcorr-1), (rcorr-1)))
        except:
            result = 0
    return int(result*1000)/1000 # round(result,2)


def cramer_matrix(data_encoded):
    """
    calculate Cramers V statistic for categorial-categorial association for a dataframe
    """
    # iterate over each column
    rows= []
    for var1 in data_encoded:
        # iterate over each column
        col = []
        for var2 in data_encoded :
            # Compute the cramer's V
            cramers =CramersCorrected(data_encoded[var1], data_encoded[var2]) # Cramer's V test
            col.append(round(cramers,2)) # Keeping of the rounded value of the Cramer's V  
        rows.append(col)
    
    # Store the result
    cramers_results = np.array(rows)
    # Convert into a dataframe
    df = pd.DataFrame(cramers_results, columns = data_encoded.columns, index =data_encoded.columns)
    return df
```

```
# This function with help you get only the categorical variable
def get_categorical_variables(df):
    """
    This function with help you get only the categorical variable
    Input:
        df: dataframe containing the data
    Output:
        col_categorical: list of column of the dataframe that are not numeric
    """
    # Get the types of each column
    df_types = df.dtypes
    # Select only the column that are object, that means, not numerical
    df_cat = df_types[df_types == "object"]
    # Transform into a list
    col_categorical = list(df_cat.index)
    return col_categorical
```

```
col_list_categorical=get_categorical_variables(df_final)
cramer_matrix(df_final[col_list_categorical])

# Is there is a relationship between the show type and the theatre: Yes, but not strong
```
